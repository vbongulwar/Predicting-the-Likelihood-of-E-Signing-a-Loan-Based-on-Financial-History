```python
# drop fields like user identifier, categorical category(pay info) and the response variable (e-sign)
dataset2 = dataset.drop(columns = ['entry_id', 'pay_schedule', 'e_signed'])
```

```python
fig = plt.figure(figsize=(15, 12)) # this size fits myscreen 
plt.suptitle('Histograms of Numerical Columns', fontsize=20)
#each columns new subplot and title it accordingly
for i in range(dataset2.shape[1]):
    plt.subplot(6, 3, i + 1)
    f = plt.gca()
    f.set_title(dataset2.columns.values[i])
#vals restrict the bins size of the histogram to 100
    vals = np.size(dataset2.iloc[:, i].unique())
    if vals >= 100:
        vals = 100
    
    plt.hist(dataset2.iloc[:, i], bins=vals, color='#b5651d')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
```

```python
#Correlation with response variable 
dataset2.corrwith(dataset.e_signed).plot.bar(
        figsize = (20, 10), title = "Correlation with E Signed", fontsize = 15,
        rot = 45, grid = True)
```

```python
sn.set(style="white")
```

```python
#compute the correlation matrix
corr = dataset2.corr()
```

```python
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
```

```python
f, ax = plt.subplots(figsize=(18, 15))
cmap = sn.diverging_palette(220, 10, as_cmap=True)
sn.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
